# Comparing different products

```python
def prepareDF(df,product_Code):
    colnames = ['PERIOD','DECLARANT','PARTNER','Value','FLOW']
    country_ID = ['0','FR','NL','DE','IT','UK','IE','DK','EL','PT','ES','BE','LU','SE','FI','AT','MT','EE','LV','LT','PL','CZ','SK','HU','RO','BG','SI','HR','CY']
    
    df['PERIOD']=df['PERIOD'].str.replace('Jan. ','01.') #cant do %b since it is german, so gonna change it to sth that pd.to_datetime can read
    df['PERIOD']=df['PERIOD'].str.replace('Feb. ','02.')
    df['PERIOD']=df['PERIOD'].str.replace('Mar. ','03.')
    df['PERIOD']=df['PERIOD'].str.replace('Apr. ','04.')
    df['PERIOD']=df['PERIOD'].str.replace('Mai. ','05.')
    df['PERIOD']=df['PERIOD'].str.replace('Jun. ','06.')
    df['PERIOD']=df['PERIOD'].str.replace('Jul. ','07.')
    df['PERIOD']=df['PERIOD'].str.replace('Aug. ','08.')
    df['PERIOD']=df['PERIOD'].str.replace('Sep. ','09.')
    df['PERIOD']=df['PERIOD'].str.replace('Okt. ','10.')
    df['PERIOD']=df['PERIOD'].str.replace('Nov. ','11.')
    df['PERIOD']=df['PERIOD'].str.replace('Dez. ','12.')
    df['PERIOD'] = pd.to_datetime(df['PERIOD'], format = "%m.%Y").dt.to_period('M')
    
    df['DECLARANT'] = pd.factorize(df.DECLARANT)[0] + 1
    df['PARTNER'] = pd.factorize(df.PARTNER)[0] + 1
    
    for x in range(1, 29):
        df['DECLARANT'] = df['DECLARANT'].replace(x ,country_ID[x])
    
    for x in range(1, 29):
        df['PARTNER'] = df['PARTNER'].replace(x ,country_ID[x])
    
    df = df.loc[df['PRODUCT'] == product_Code]
    
    df=df.filter(items=colnames)
    
    return df
```

```python
def pivotDF(df, einfuhr):
    
    if einfuhr:
        df = df.loc[df['FLOW'] == 'EINFUHR'].copy()
        df = df.pivot_table(values='Value', index=['PERIOD','DECLARANT'], columns='PARTNER', aggfunc="last")
    else:
        df = df.loc[df['FLOW'] == 'AUSFUHR'].copy()
        df = df.pivot_table(values='Value', index=['PERIOD','PARTNER'], columns='DECLARANT', aggfunc='first')
    
    return df
```

```python
def testSimilarity(df):
    count1 = len(df.index)
    print("DataFrame has " + str(count1) + " rows")
    dfImport = df.loc[df['FLOW'] == 'EINFUHR'].copy()
    dfExport = df.loc[df['FLOW'] == 'AUSFUHR'].copy()
    dfSimilar = pd.merge(dfImport, dfExport, how='inner', left_on=['PERIOD','DECLARANT','PARTNER','Value'], right_on=['PERIOD','PARTNER','DECLARANT','Value'])
    count2 = len(dfSimilar.index)
    print("DataFrame has " + str(count2) + " matching values")
    
    possibleFraud = (1-(count2/count1))*100
    print("According to that " + str(possibleFraud) + "% would be fraudulent")
    
    dfDeviation = pd.merge(dfImport, dfExport, how='inner', left_on=['PERIOD','DECLARANT','PARTNER'], right_on=['PERIOD','PARTNER','DECLARANT'])
    nullCount1 = len(dfDeviation.index)
    dfDeviation = dfDeviation[(dfDeviation['Value_x'] != ':') & (dfDeviation['Value_y'] != ':')]
    nullCount2 = len(dfDeviation.index)
    nullPercent = (nullCount2/nullCount1)*100
    print(str(nullPercent) +'% of the data is Null')
    #For mean deviation
    dfDeviation['Value_x'] = dfDeviation['Value_x'].str.replace('.', '')
    dfDeviation['Value_x'] = dfDeviation['Value_x'].str.replace(',', '.')
    dfDeviation['Value_x'] = dfDeviation['Value_x'].astype(float) 
    dfDeviation['Value_y'] = dfDeviation['Value_y'].str.replace('.', '')
    dfDeviation['Value_y'] = dfDeviation['Value_y'].str.replace(',', '.')
    dfDeviation['Value_y'] = dfDeviation['Value_y'].astype(float)
    dfDeviation['deviation'] =1-(dfDeviation.loc[:, ['Value_x', 'Value_y']].min(axis=1)/dfDeviation.loc[:, ['Value_x', 'Value_y']].max(axis=1))
    
    meanDeviation = (dfDeviation['deviation'].mean())*100
    print('The mean deviation is about ' + str(meanDeviation) + '%')
```
